```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from keras.layers import Flatten, Dense
from keras.models import Model
from keras.preprocessing.image import ImageDataGenerator #, load_img
from tensorflow.keras.utils import img_to_array,load_img
from keras.applications.mobilenet import MobileNet, preprocess_input 
from keras.losses import categorical_crossentropy
```

#  Building our Model To train the data 

```python
# Working with pre trained model 

base_model = MobileNet( input_shape=(224,224,3), include_top= False )

for layer in base_model.layers:
  layer.trainable = False


x = Flatten()(base_model.output)
x = Dense(units=2 , activation='softmax' )(x)

# creating our model.
model = Model(base_model.input, x)
```

```python
model.compile(optimizer='adam', loss= categorical_crossentropy , metrics=['accuracy']  )
```

# Preparing our data using data generator

```python
train_datagen = ImageDataGenerator(
     zoom_range = 0.2, 
     shear_range = 0.2, 
     horizontal_flip=True, 
     rescale = 1./255
)

train_data = train_datagen.flow_from_directory(directory= "C:\Users\THANOS\Desktop\Shriram Project\Emotion-detection-main\Emotion-detection-main\train", 
                                               target_size=(224,224), 
                                               batch_size=32,
                                  )


train_data.class_indices
```

```python
val_datagen = ImageDataGenerator(rescale = 1./255 )

val_data = val_datagen.flow_from_directory(directory= "C:\Users\THANOS\Desktop\Shriram Project\Emotion-detection-main\Emotion-detection-main\test", 
                                           target_size=(224,224), 
                                           batch_size=32,
                                  )
```

# visualizaing the data that is fed to train data gen

```python
# to visualize the images in the traing data denerator 

t_img , label = train_data.next()

#-----------------------------------------------------------------------------
# function when called will prot the images 
def plotImages(img_arr, label):
  """
  input  :- images array 
  output :- plots the images 
  """
  count = 0
  for im, l in zip(img_arr,label) :
    plt.imshow(im)
    plt.title(im.shape)
    plt.axis = False
    plt.show()
    
    count += 1
    if count == 10:
      break

#-----------------------------------------------------------------------------
# function call to plot the images 
plotImages(t_img, label)
```

# having early stopping and model check point

```python
## having early stopping and model check point 

from keras.callbacks import ModelCheckpoint, EarlyStopping

# early stopping
es = EarlyStopping(monitor='val_accuracy', min_delta= 0.01 , patience= 5, verbose= 1, mode='auto')

# model check point
mc = ModelCheckpoint(filepath="best_model3.h5", monitor= 'val_accuracy', verbose= 1, save_best_only= True, mode = 'auto')

# puting call back in a list 
call_back = [es, mc]
```

```python
steps_per_epoch = len(train_data)//10

validation_steps = len(val_data)//10
steps_per_epoch,validation_steps
```

```python
hist = model.fit_generator(train_data, 
                           steps_per_epoch= 4, 
                           epochs= 10, 
                           validation_data= val_data, 
                           validation_steps= 2, 
                           callbacks=[es,mc])
```
